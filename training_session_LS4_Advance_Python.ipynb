#2&3 ) Program Perhitungan BMI Sobat Startup Campus 5
Buatlah program untuk menghitung BMI!

```python
def bmi(berat, tinggi):
    tinggi_dalam_meter = tinggi / 100
    bmi = berat / (tinggi_dalam_meter ** 2)

    return bmi

berat = float(input ("Masukan berat badan kalian :"))
tinggi = float(input ("Masukan tinggi badan kalian :"))

hasil = bmi(berat,tinggi)

print("Hasil BMI kalian :", hasil)

if hasil >= 18.5 and hasil < 24.9:
  print ("berat badan kamu Normal")
elif hasil >= 25.0 and hasil < 29.9:
  print ("berat badan kamu Overweight")
elif hasil >= 30.0 and hasil < 34.9:
  print ("berat badan kamu Obesity")
elif hasil < 18.5:
  print ("berat badan kamu Underweight")

```
